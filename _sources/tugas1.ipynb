```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances
data = pd.read_csv('https://raw.githubusercontent.com/nurulfirdausy/Penambangan-Data/master/data_distance.csv')
data
```

```python
X = data[['jenis kelamin', 'umur', 'status', 'jabatan']].copy()
X['jenis kelamin'] = pd.Categorical(X['jenis kelamin']).codes
X['umur'] = pd.Categorical(X['umur']).codes
X['status'] = pd.Categorical(X['status']).codes
X['jabatan'] = pd.Categorical(X['jabatan'], categories=['it support', 'supervisor'], ordered=True).codes
```

```python
dist_matrix = pairwise_distances(X, metric='manhattan') / max(pairwise_distances(X, metric='manhattan').flatten())
```
